## Transformer

### -Question answering

```python

from transformers import pipeline
question_answering = pipeline("question-answering")
context = """
        Le traitement naturel du langage est une technologie permettant aux machines de comprendre le langage humain grâce à l’intelligence artificielle. Découvrez tout ce que vous devez savoir à ce sujet.
    """
question = "Qu'est ce le traitement naturel du langage ?"
result = question_answering(question=question, context=context)

print("Reponse: ", result['answer'])
```

### -Translation

```python
# English to French
translator_en_fr = pipeline("translation_en_to_fr")
print("French: ", translator_en_fr("This is an example sentence.", max_length=50)[0]['translation_text'])

```

### -Filling masked text

```python
from transformers import pipeline

nlp = pipeline("fill-mask")

from pprint import pprint

res = nlp(f"Pour se {nlp.tokenizer.mask_token} il faut aller à l'hopital.")
pprint()
```

```python
from transformers import pipeline

# French
TXT_GEN_FR = pipeline('text-generation', model='dbddv01/gpt2-french-small')

print(TXT_GEN_FR("je vais ", max_length=50, do_sample=False))
```
